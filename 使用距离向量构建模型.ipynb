```python
from sklearn.datasets import make_blobs
import numpy as np
blobs, classes = make_blobs(500, centers=3)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(blobs)
kmeans.inertia_
```

```python
silhouette_samples.mean()
```

```python
blobs, classes = make_blobs(500, centers=10)
sillhouette_avgs = []
for k in range(2, 60):
    kmean = KMeans(n_clusters=k).fit(blobs)
    sillhouette_avgs.append(metrics.silhouette_score(blobs,kmean.labels_))
f, ax = plt.subplots(figsize=(10,10))
ax.plot(sillhouette_avgs)
```

```python
from sklearn.datasets import make_blobs
X, labels = make_blobs(100, centers=1)
import numpy as np
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=1)
kmeans.fit(X)

```

```python
f,ax = plt.subplots(figsize=(7.5,7.5))
ax.set_title("Blob")
ax.scatter(X[:, 0], X[:, 1], label='Points', color='b')
ax.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='red',label='Centroid')
ax.legend()
```

```python
distances = kmeans.transform(X)
sorted_idx = np.argsort(distances.ravel())[::-1][:5]
```

```python
sorted_idx
```

```python
f, ax = plt.subplots(figsize=(7.5,7.5))
ax.set_title("Single Cluster")
ax.scatter(X[:,0], X[:,1], color='blue', label = 'Points')
ax.scatter(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[:, 1],label='Centroid', color='r')
ax.scatter(X[sorted_idx][:,0], X[sorted_idx][:,1], label='Extreme Value', edgecolors='g', facecolors='none', s=100)
ax.legend()
```

```python
new_X = np.delete(X, sorted_idx, axis=0)
new_kmeans = KMeans(n_clusters=1)
new_kmeans.fit(new_X)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
ax.set_title("Extreme Values Removed")
ax.scatter(new_X[:, 0], new_X[:, 1], label='Pruned Points')
ax.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='red',label='Old Centroid', s=80, alpha=.5)
ax.scatter(new_kmeans.cluster_centers_[:, 0],new_kmeans.cluster_centers_[:, 1], label='New Centroid',color='m', s=80, alpha=.5)
ax.legend(loc='best')
```

```python
from scipy import stats
emp_dist = stats.multivariate_normal(kmeans.cluster_centers_.ravel())
lowest_prob_idx = np.argsort(emp_dist.pdf(X))[:5]
np.all(X[sorted_idx] == X[lowest_prob_idx])
```

```python
from sklearn import datasets
import numpy as np
iris = datasets.load_iris()
X,y = iris.data, iris.target
iris.feature_names
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X[:,:2],X[:,3])
print("The MSE is: {:.2}".format(np.power(X[:,3] - lr.predict(X[:,:2]),2).mean()))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knnr = KNeighborsRegressor(n_neighbors=10)
knnr.fit(X[:,:2],X[:,3])
print("The MSE is: {:.2}".format(np.power(X[:,3] - knnr.predict(X[:,:2]),2).mean()))
```

```python
f, ax = plt.subplots(nrows=2, figsize=(7, 10))
ax[0].set_title("Predictions")
ax[0].scatter(X[:, 0], X[:, 1], s=lr.predict(X[:,:2])*80, label='LR Predictions', color='c', edgecolors='black')
ax[1].scatter(X[:, 0], X[:, 1], s=knnr.predict(X[:,:2])*80, label='k-NN Predictions', color='m', edgecolors='black')
ax[0].legend()
ax[1].legend()
```
